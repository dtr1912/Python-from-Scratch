## collections.Counter()
A counter is a container that stores elements as dictionary keys, and their counts are stored as dictionary values.


```python
from collections import Counter
X = int(input())
sizes = list(map(int, input().split()))
money = 0
N = int(input())
for _ in range(N):
    lst = list(map(int, input().split()))
    cnt = Counter(sizes)
    if lst[0] in cnt.keys():
        money += lst[1]
        sizes.remove(lst[0])
    else:
        money += 0
print(money)
```

## collections.namedtuple()


```python
from collections import namedtuple
N = int(input())
headers = input()
# namedtuple giúp tạo ra một lớp Student có các thuộc tính tương ứng với headers mà không cần viết một lớp hoàn chỉnh.
Students = namedtuple('Students', headers)
total_marks = 0
for i in range(N):
    student_data = input().split()
    # *student_data giúp truyền từng phần tử của student_data vào các thuộc tính của Student.
    std = Students(*student_data)
    #int(student.MARKS) chuyển đổi điểm từ chuỗi sang số nguyên để tính tổng điểm.
    total_marks += int(std.MARKS)
avg = total_marks/N
print(f"{avg:.2f}")
```

## Strings


```python
def minion_game(string):
    vowels = 'AEOUI'
    N = len(string)
    score_k = 0
    score_s = 0
    for i in range(N):
        if string[i] in vowels:
            score_k += N - i
        else:
            score_s += N - i
    if score_k > score_s:
        print(f"Kevin {score_k}")
    elif score_k < score_s:
        print(f"Stuart {score_s}")
    else:
        print('Draw')

if __name__ == '__main__':
    s = input()
    minion_game(s)
```

## Find a string


```python
def count_substring(string, sub_string):
    # đếm số lần xuất hiện của sub_string trong string.
    count = 0
    # vị trí bắt đầu tìm kiếm trong chuỗi string.
    start = 0 
    # Vòng lặp vô hạn này sẽ tiếp tục cho đến khi chúng ta tìm thấy tất cả các xuất hiện của sub_string.
    while True:
        start = string.find(sub_string, start) + 1
        if start > 0:
            count += 1
        else:
            break
    return count

if __name__ == '__main__':
    string = input().strip()
    sub_string = input().strip()
    
    count = count_substring(string, sub_string)
    print(count)
```

- Phương thức find() trong Python là một phương thức của chuỗi (str) dùng để tìm vị trí xuất hiện đầu tiên của một chuỗi con trong chuỗi gốc. Nếu chuỗi con không được tìm thấy, phương thức này sẽ trả về -1.
- Syntax: str.find(sub[, start[, end]])
- Tham số
  - sub: Chuỗi con cần tìm
  - start (tuỳ chọn): Vị trí bắt đầu tìm kiếm. Mặc định là 0.
  - end (tuỳ chọn): Vị trí kết thúc tìm kiếm. Mặc định là độ dài của chuỗi gốc.
- Giá trị trả về: Chỉ số (index) của vị trí bắt đầu của chuỗi con đầu tiên trong chuỗi gốc. Nếu không tìm thấy, trả về -1.

## Nested List
Given the names and grades for each student in a class of N students, store them in a nested list and print the name(s) of any student(s) having the second lowest grade.

Note: If there are multiple students with the second lowest grade, order their names alphabetically and print each name on a new line.


```python
if __name__ == '__main__':
    lst = []
    for _ in range(int(input("Enter the number of students: "))):
        name = input("Enter the name of the student: ")
        score = map(float, input("Enter the score of the student: "))
        lst.append([name, score])

    # Sắp xếp danh sách theo điểm số, rồi theo tên
    lst.sort(key=lambda x: (x[1], x[0]))

    # Tìm điểm số thấp nhất
    lowest_score = lst[0][1]

    # Tìm điểm số thấp thứ hai
    second_lowest_score = None
    for score in lst:
        if score[1] > lowest_score:
            second_lowest_score = score[1]
            break

    # Tạo danh sách các học sinh có điểm số thấp thứ hai
    students_with_second_lowest_score = [student[0] for student in lst if student[1] == second_lowest_score]

    # In tên các học sinh có điểm số thấp thứ hai
    for student in students_with_second_lowest_score:
        print(student)
```

## Map and Lambda Function 



```python
cube = lambda x: x**3
def fibonacci(n):
    if n <= 0:
        return []
    elif n == 1:
        return [0]
    f_list = [0, 1]
    for i in range(2, n):
        # Phần tử tiếp theo bằng 2 phần tử cuối (2 phần tử ngay trước nó) list cộng lại 
        next_f =  f_list[-2] + f_list[-1] 
        f_list.append(next_f)
    return f_list
if __name__ == '__main__':
    n = int(input())
    print(list(map(cube, fibonacci(n))))
```
