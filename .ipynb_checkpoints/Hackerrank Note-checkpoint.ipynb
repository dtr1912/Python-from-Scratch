## collections.Counter()
A counter is a container that stores elements as dictionary keys, and their counts are stored as dictionary values.


```python
from collections import Counter
X = int(input())
sizes = list(map(int, input().split()))
money = 0
N = int(input())
for _ in range(N):
    lst = list(map(int, input().split()))
    cnt = Counter(sizes)
    if lst[0] in cnt.keys():
        money += lst[1]
        sizes.remove(lst[0])
    else:
        money += 0
print(money)
```

## collections.namedtuple()


```python
from collections import namedtuple
N = int(input())
headers = input()
# namedtuple gip to ra mt lp Student c cc thuc tnh tng ng vi headers m khng cn vit mt lp hon chnh.
Students = namedtuple('Students', headers)
total_marks = 0
for i in range(N):
    student_data = input().split()
    # *student_data gip truyn tng phn t ca student_data vo cc thuc tnh ca Student.
    std = Students(*student_data)
    #int(student.MARKS) chuyn i im t chui sang s nguyn  tnh tng im.
    total_marks += int(std.MARKS)
avg = total_marks/N
print(f"{avg:.2f}")
```

## Strings


```python
def minion_game(string):
    vowels = 'AEOUI'
    N = len(string)
    score_k = 0
    score_s = 0
    for i in range(N):
        if string[i] in vowels:
            score_k += N - i
        else:
            score_s += N - i
    if score_k > score_s:
        print(f"Kevin {score_k}")
    elif score_k < score_s:
        print(f"Stuart {score_s}")
    else:
        print('Draw')

if __name__ == '__main__':
    s = input()
    minion_game(s)
```

## Find a string


```python
def count_substring(string, sub_string):
    # m s ln xut hin ca sub_string trong string.
    count = 0
    # v tr bt u tm kim trong chui string.
    start = 0 
    # Vng lp v hn ny s tip tc cho n khi chng ta tm thy tt c cc xut hin ca sub_string.
    while True:
        start = string.find(sub_string, start) + 1
        if start > 0:
            count += 1
        else:
            break
    return count

if __name__ == '__main__':
    string = input().strip()
    sub_string = input().strip()
    
    count = count_substring(string, sub_string)
    print(count)
```

- Phng thc find() trong Python l mt phng thc ca chui (str) dng  tm v tr xut hin u tin ca mt chui con trong chui gc. Nu chui con khng c tm thy, phng thc ny s tr v -1.
- Syntax: str.find(sub[, start[, end]])
- Tham s
  - sub: Chui con cn tm
  - start (tu chn): V tr bt u tm kim. Mc nh l 0.
  - end (tu chn): V tr kt thc tm kim. Mc nh l  di ca chui gc.
- Gi tr tr v: Ch s (index) ca v tr bt u ca chui con u tin trong chui gc. Nu khng tm thy, tr v -1.

## Nested List
Given the names and grades for each student in a class of N students, store them in a nested list and print the name(s) of any student(s) having the second lowest grade.

Note: If there are multiple students with the second lowest grade, order their names alphabetically and print each name on a new line.


```python
if __name__ == '__main__':
    lst = []
    for _ in range(int(input("Enter the number of students: "))):
        name = input("Enter the name of the student: ")
        score = map(float, input("Enter the score of the student: "))
        lst.append([name, score])

    # Sp xp danh sch theo im s, ri theo tn
    lst.sort(key=lambda x: (x[1], x[0]))

    # Tm im s thp nht
    lowest_score = lst[0][1]

    # Tm im s thp th hai
    second_lowest_score = None
    for score in lst:
        if score[1] > lowest_score:
            second_lowest_score = score[1]
            break

    # To danh sch cc hc sinh c im s thp th hai
    students_with_second_lowest_score = [student[0] for student in lst if student[1] == second_lowest_score]

    # In tn cc hc sinh c im s thp th hai
    for student in students_with_second_lowest_score:
        print(student)
```

## Map and Lambda Function 



```python
cube = lambda x: x**3
def fibonacci(n):
    if n <= 0:
        return []
    elif n == 1:
        return [0]
    f_list = [0, 1]
    for i in range(2, n):
        # Phn t tip theo bng 2 phn t cui (2 phn t ngay trc n) list cng li 
        next_f =  f_list[-2] + f_list[-1] 
        f_list.append(next_f)
    return f_list
if __name__ == '__main__':
    n = int(input())
    print(list(map(cube, fibonacci(n))))
```
