Trong matplotlib,

## Các điều chỉnh trong Plot: Line Colors and Styles


```python
plt.plot(x, np.sin(x - 0), color='blue')        # specify color by name
plt.plot(x, np.sin(x - 1), color='g')           # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75')        # Grayscale between 0 and 1
plt.plot(x, np.sin(x - 3), color='#FFDD44')     # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB tuple, values 0 to 1
plt.plot(x, np.sin(x - 5), color='chartreuse'); # all HTML color names supported
```

```python
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');

# For short, you can use the following codes:
plt.plot(x, x + 4, linestyle='-')  # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':');  # dotted
```

# <span style="color: red;">3. Visualizing Errors</span>

## Basic Errorbars

```python
x = np.linspace(0,10,50)
dy =0.8
y= np.sin(x) + dy*np.random.rand(50)
# fmt='.k' có nghĩa là dùng điểm (.) và màu đen (k)
plt.errorbar(x,y ,yerr= dy, fmt ='.k')
```

'fmt' là mã định dạng kiểm soát sự xuất hiện của các đường và điểm và có cú pháp giống như ký hiệu viết tắt được sử dụng trong plt.plot

```python

plt.errorbar(x, y, yerr = dy, fmt='o', color ='black',
             ecolor = 'lightgray', elinewidth = 3, capsize =0)
```

- x: Giá trị trên trục x.
- y: Giá trị trên trục y.
- yerr=dy: Sai số của các giá trị y, được biểu diễn dưới dạng các thanh lỗi.
- fmt='o': Định dạng của điểm trên biểu đồ. Ở đây, 'o' biểu diễn các điểm tròn (marker).
- color='black': Màu của các điểm tròn (markers).
- ecolor='lightgray': Màu của các thanh lỗi (error bars).
- elinewidth=3: Độ rộng của các thanh lỗi.
- capsize=0: Độ dài của các mũ (caps) ở hai đầu thanh lỗi. Ở đây, giá trị 0 có nghĩa là không có mũ.

Ngoài các lựa chọn này ra còn có thể chọn 'xerr' là các errorbars nằm ngang,  thanh lỗi một phía và nhiều biến thể khác

## Continuous Errors

Trong một vài trường hợp, chúng ta mong muốn biểu diễn errorbars như một đại lượng liên tục.
Sử dụng plt.plot và plt.fill_between 

Ví dụ : Thực hiện Gaussian process regression, sử dụng Scikit-Learn API.

```python
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, ConstantKernel as C
# Xác định model và vẽ vài dữ liệu
model = lambda x: x*np.sin(x)
xdata = np.array([1,3 ,5,6,8])
ydata = model(xdata)
# Compute the Gaussian process fit
kernel = C(1.0, (1e-4, 23.0)) * RBF(1.0, (1e-4, 1e1))
gp =  GaussianProcessRegressor(kernel=kernel, n_restarts_optimizer=100)
gp.fit(xdata[:,np.newaxis],ydata)
xfit = np.linspace(0,10,1000)
yfit = gp.predict(xfit[:, np.newaxis])
ytrue = model(xfit.flatten())  # True values corresponding to xfit
MSE = np.mean((ytrue - yfit)**2)
dyfit = 2 * np.sqrt(MSE)
```

```python
plt.plot(xdata, ydata, 'or')
plt.plot(xfit, yfit, '-g')
# plt.fill_between sử dụng để tạo vùng tô màu giữa hai dãy dữ liệu
plt.fill_between(xfit, yfit - dyfit, yfit + dyfit,
                 color='gray', alpha=0.2)
plt.xlim(0, 10);
```

# <span style="color: red;">4. Contour Plots</span>

- Contour Plot (Biểu đồ đường đồng mức):
Contour Plot biểu thị dữ liệu 3 chiều trên một đồ thị 2 chiều bằng cách vẽ các đường đồng mức (contour lines) theo một chiều thứ ba. Đây là công cụ hữu ích để hình dung mối quan hệ giữa ba biến.

## Visualizing a Three-Dimensional Function

```python
# z=f(x,y)
 
def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)
```

plt.contour nhận ba đối số: lưới các giá trị x , lưới các giá trị y và lưới các giá trị z . Các giá trị x và y biểu thị các vị trí trên biểu đồ và các giá trị z sẽ được biểu thị bằng các contour level.
Có lẽ cách đơn giản nhất để chuẩn bị dữ liệu đó là sử dụng hàm np.meshgrid xây dựng lưới hai chiều từ mảng một chiều:

```python
x = np.linspace(0,5,50)
y = np.linspace(0,5,40)
X, Y = np.meshgrid(x,y)
Z = f(X,Y)
#
plt.contour(X,Y,Z, colors ='black')
```

Lưu ý rằng theo mặc định khi sử dụng một màu duy nhất, các giá trị âm được biểu thị bằng đường đứt nét và giá trị dương bằng đường liền nét. Ngoài ra, các dòng có thể được mã hóa màu bằng cách chỉ định bản đồ màu với đối số cmap. Ở đây,  chỉ định rằng  muốn vẽ nhiều đường hơn — 20 khoảng cách đều nhau trong phạm vi dữ liệu:

```python
# RdGy (short for Red-Gray) colormap
plt.contour(X,Y,Z, 20, cmap ='RdGy')
```

```python
# switching to a filled contour plot using the plt.contourf() function
plt.contourf(X,Y,Z, 20, cmap ='RdGy')
# Tự động tạo một trục với thông tin màu được gán nhãn
plt.colorbar();
```

Thay vì dựa vào các đường đồng mức rời rạc với plt.contourf(), plt.imshow() giải thích một lưới dữ liệu hai chiều trực tiếp như một hình ảnh. Phương pháp này cung cấp một phép ánh xạ màu liên tục trên toàn bộ biểu đồ, tránh sự xuất hiện như các bước nhưng màu sắc có thể xảy ra với biểu đồ đường đồng mức.

```python
plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower',
           cmap='RdGy')
plt.colorbar()
```

Các vấn đề tiềm ẩn khi sử dụng plt.imshow():
- Xác định phạm vi (extent) một cách thủ công: extent [xmin, xmax, ymin, ymax] để xác định phạm vi không gian của hình ảnh trên đồ thị.
- Mặc định, plt.imshow() tuân theo định nghĩa mảng hình ảnh tiêu chuẩn, trong đó gốc là ở góc trên bên trái. Điều này khác với hầu hết các đồ thị đường viền trong Matplotlib, nơi gốc được đặt ở góc dưới bên trái. Sử dụng origin='lower' trong plt.imshow() để xác định gốc dưới bên trái để tránh sự nhầm lẫn và hiển thị không chính xác trên đồ thị.
- plt.imshow() sẽ tự động điều chỉnh tỷ lệ khung hình của trục để phù hợp với kích thước dữ liệu đầu vào. Điều này có thể được thay đổi bằng cách thiết lập plt.axis() để làm cho các đơn vị trên trục x và y phù hợp.

Đôi khi có thể hữu ích khi kết hợp các contour plots và image plots. Ví dụ: ở đây chúng ta sẽ sử dụng hình nền trong suốt một phần (với độ trong suốt được đặt thông qua tham số alpha) và vẽ chồng lên các đường viền bằng nhãn trên chính đường viền đó (sử dụng plt.clabel()):

```python
contours = plt.contour(X, Y, Z, 3, colors='black')
plt.clabel(contours, inline=True, fontsize=8)

plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower',
           cmap='RdGy', alpha=0.5)
plt.colorbar();
```

# <span style="color: red;">5. Histograms, Binnings, and Density </span>

```python
%matplotlib inline
plt.style.use('seaborn-white')

data = np.random.randn(1000)
```

```python
plt.hist(data)
```

```python
plt.hist(data, bins =30, density =True, alpha =0.5,
        histtype ='stepfilled', color = 'steelblue',
        edgecolor ='none')
```

- data: Là dữ liệu của bạn, là một mảng hoặc tập hợp dữ liệu bạn muốn vẽ histogram.

- bins: Số lượng các khoảng (bins) mà dữ liệu sẽ được chia thành. Với bins=30, bạn sẽ có 30 khoảng trong histogram.

- density=True (trước đây là normed=True): Tham số này cho biết liệu histogram có nên được chuẩn hóa để tạo thành mật độ xác suất hay không. Nó tỷ lệ chiều cao của các bin sao cho tổng diện tích dưới histogram là 1.

- alpha=0.5: Điều này điều khiển độ mờ của các thanh histogram. Giá trị alpha là 0.5 làm cho các thanh trở nên mờ một nửa.

- histtype='stepfilled': Tham số này xác định loại histogram. 'stepfilled' sẽ vẽ đường nối các đỉnh của các thanh histogram và tô màu vùng dưới đường.

- color='steelblue': Đặt màu của các thanh histogram thành 'steelblue'.

- edgecolor='none': Loại bỏ màu viền của các thanh histogram, làm cho chúng hòa vào nền của biểu đồ.
